```python
import numpy as np
emotion_labels=['anger', 'disgust', 'fear', 'joy', 'sadness', 'surprise', 'neutral']
isbn=[]
emotion_scores=[]

def calculate_max_emotion_scores(predictions):
    per_emotion_scores = {label: [] for label in emotion_labels}
    
    for sentence_prediction in predictions:
        for pred in sentence_prediction:
            label = pred['label']
            score = pred['score']
            if label in per_emotion_scores:
                per_emotion_scores[label].append(score)
    return {label: max(scores) if scores else 0 for label, scores in per_emotion_scores.items()}
```

```python
# Initialize emotion_scores as a dictionary of lists if not already done
if not isinstance(emotion_scores, dict):
    emotion_scores = {label: [] for label in emotion_labels}

for i in range(10):
    isbn.append(books["isbn13"][i])
    # Split into sentences and remove empty strings
    sentences = [s for s in books["description"][i].split(".") if s.strip()]
    predictions = classifier(sentences)
    if isinstance(predictions, list) and predictions and isinstance(predictions[0], dict):
        predictions = [predictions]
    max_scores = calculate_max_emotion_scores(predictions)
    for label in emotion_labels:
        emotion_scores[label].append(max_scores[label])
```

```python
from tqdm import tqdm

emotion_labels = ["anger", "disgust", "fear", "joy", "sadness", "surprise", "neutral"]
isbn = []
emotion_scores = {label: [] for label in emotion_labels}

for i in tqdm(range(len(books))):
    isbn.append(books["isbn13"][i])
    sentences = books["description"][i].split(".")
    predictions = classifier(sentences)
    if isinstance(predictions, list) and predictions and isinstance(predictions[0], dict):
        predictions = [predictions]
    max_scores = calculate_max_emotion_scores(predictions)
    for label in emotion_labels:
        emotion_scores[label].append(max_scores[label])
```

```python
emotions_df = pd.DataFrame(emotion_scores)
emotions_df["isbn13"] = isbn
emotions_df
```

```python
books = pd.merge(books, emotions_df, on = "isbn13")
books
```

```python
books.to_csv("books_with_emotions.csv", index = False)
```
